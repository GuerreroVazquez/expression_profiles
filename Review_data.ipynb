```python
expression_data = pd.read_csv('data/RNAseq_z_score_adjustedCombat.csv')
```

```python
ensembl_to_symbol = pd.read_csv('data/ensmbl_list.csv', index_col = 0)
ensembl_to_symbol.head()
```

```python
ensembl_to_symbol_dic = ensembl_to_symbol.to_dict('index')
```

```python
ensembl_to_symbol_dic
```

```python
ensembl_genes = expression_data.columns[1:-2]
```

```python
ensembl_genes= [gene.split('.')[0] for gene in ensembl_genes ]
```

```python
ensembl_genes[-10:]
```

```python
column_symbols = [ensembl_to_symbol_dic[ensemble]['geneSymbol'] for ensemble in ensembl_genes]
column_symbols[:10]
```

```python
column_symbols_ = ['Sample']
column_symbols_.extend(column_symbols)

column_symbols_.extend(["Age", "Experiment"])
```

```python
len(column_symbols_)
```

```python
expression_data.columns = column_symbols_
expression_data.head()
```
